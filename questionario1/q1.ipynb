#Exerc√≠cio2

```python
df = pd.read_csv("data/iris-with-errors.csv", header=(0))
df
```

```python
for coluna in df.columns:
    if coluna != 'species':
        df[coluna] = pd.to_numeric(df[coluna], errors='coerce')
df
```

```python
df = df.fillna(df.mean(numeric_only=True))
df
```

```python
resultado_final = df.median(numeric_only=True)
medianas = resultado_final.round(2).tolist()
medianas
```

```python
print(f"Resposta da quest√£o 2: Ap√≥s a limpeza dos dados, as medianas de cada atributo s√£o {medianas}")
```

#Exerc√≠cio3

```python
df = pd.read_csv("data/iris.csv", header=(0))
df
```

```python
plt.figure_size = (10,10)
for atribute in ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']:
    df.boxplot(column=[atribute], by='species')
    plt.title('Boxplot dos Atributos da Iris', fontsize=16)
    plt.ylabel('Valores (cm)', fontsize=12)
    plt.show()
```

```python
print(f"Resposta da quest√£o 3: Todos os atributos possuem outliers")
```

#Exerc√≠cio4

```python
df = pd.read_csv("data/Advertising.csv", header=(0))
df
```

```python
correlacao = df.corr()
correlacao
```

```python
correlacao = correlacao.unstack()
correlacao
```

```python
correlacao = correlacao.sort_values(kind="quicksort", ascending=False)
correlacao
```

```python
correlacao = correlacao[correlacao < 1.0]
correlacao
```

```python
best_correlation = correlacao.head(1)
best_correlation = str(best_correlation).split()[:2]
best_correlation
```

```python
print(f"Resposta da quest√£o 4: Os atributos com maior correla√ß√£o s√£o {best_correlation[0]} e {best_correlation[1]}")
```
