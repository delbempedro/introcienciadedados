#Exercício2

```python
df = pd.read_csv("data/iris-with-errors.csv", header=(0))
df
```

```python
df = df.replace('?', np.nan)
df
```

```python
for coluna in df.columns:
  if coluna == 'species':
    continue
  df[coluna] = pd.to_numeric(df[coluna], errors='coerce')
df
```

```python
df = df.fillna(df.mean())
for coluna in df.columns:
    if coluna == 'species':
        continue
    df[coluna] = df[coluna].fillna(df.mean())
df
```

```python
medias = []
for atributo in df.columns:
    medias.append(df[atributo].mean())
medias
```
